```python
import numpy as np
import pandas as pd
data=pd.read_csv("C:\\Users\\gp\\Downloads\\0000000000002429_training_titanic_x_y_train.csv",delimiter=",")



data=data.drop(['Name','Ticket','Fare','Pclass',"SibSp","Parch","Cabin","Embarked"],axis=1)
h=pd.get_dummies(data.Sex)
h=pd.get_dummies(data.Sex)["female"]
print(h)
h=pd.DataFrame(h)
data["sex"]=h
data=data.drop(['Sex'],1)
h=data["Age"].median()
aging=data["Age"].fillna(h)
aging=pd.DataFrame(aging)
aging
data["age"]=aging
data
data=data.drop(['Age'],1)

y=data["Survived"]
data=data.drop(["Survived"],1)
data

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(data)
```

```python
Y_train=np.array(y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(penalty="l2",C=1,multi_class="multinomial",solver='lbfgs',tol=0.1)
```

```python
model.fit(X_train,Y_train)
```

```python
data_test=pd.read_csv("C:\\Users\gp\\Downloads\\0000000000002429_test_titanic_x_test.csv",delimiter=",")
data_test
data_test=data_test.drop(['Name','Ticket','Fare','Pclass',"SibSp","Parch","Cabin","Embarked"],axis=1)
h=pd.get_dummies(data_test.Sex)
h=pd.get_dummies(data_test.Sex)["female"]
data_test["sex"]=h
h=data_test["Age"].median()
aging=data_test["Age"].fillna(h)
data_test["age"]=aging
data_test=data_test.drop(['Age','Sex'],axis=1)
X_test=sc.fit_transform(data_test)
X_test
```
